```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv('BL-Flickr-Images-Book.csv')
```

```python
df.head()
```

```python
#Part 1


def clean_title(title):
    # Remove bracket and its content
    title = re.sub(r'\[.*?\]', '', title)
    # Remove ellipsis and everything after it
    title = re.sub(r'\s*\.{3}.*', '', title)
    # Remove period and everything after it
    title = re.sub(r'\s*\..*', '', title)
    return title.strip()

# Apply the cleaning function to the Title column
df['Title'] = df['Title'].apply(clean_title)
```

```python
df
```

```python
#Part 2


df['Author'] = df['Author'].str.split(',').str[0]
df['Author'] = df['Author'].str.split('(').str[0]
df['Author'] = df['Author'].str.title()

df['Contributors'] = df['Contributors'].str.split(',').str[0]
df['Contributors'] = df['Contributors'].str.split('(').str[0]
df['Contributors'] = df['Contributors'].str.title()

df['Title'] = df['Title'].apply(lambda x: re.sub(r'\(.*\)', '', x))
df['Title'] = df['Title'].str.split('...').str[0]
df['Title'] = df['Title'].str.split('.').str[0]
```
