```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sp500 = pd.read_csv('sp500.csv')
estate = pd.read_csv('Sacramentorealestatetransactions.csv')
```

```python
sp500.head()
```

```python
sp500 = sp500[sp500['Date'] >= '2018-01-02']
```

```python
sp500.tail()
```

```python
sp500['Gain/Loss'] = sp500['Close'] - sp500['Open']
```

```python
sp500.plot(x='Date', y='Gain/Loss', title='S&P 500 High & Low Price', figsize=(20, 10))


plt.annotate('Highest Gain', xy=(sp500['Date'][sp500['Gain/Loss'].idxmax()], sp500['Gain/Loss'].max()), xytext=(sp500['Date'][sp500['Gain/Loss'].idxmax()+5], sp500['Gain/Loss'].max() + 1),
arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('Highest Gain', xy=(sp500['Date'][sp500['Gain/Loss'].idxmin()], sp500['Gain/Loss'].min()), xytext=(sp500['Date'][sp500['Gain/Loss'].idxmin()+5], sp500['Gain/Loss'].min() + 1),
arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```
